```python
# extend()
lst1 = ['Harish','Lalit',54]
print(lst1)
lst1.extend([21,True,56.9])
print(lst1)
```

```python
# Removal of items from a list:

# remove() function
df = [24, 89, "Ikshit", 7-8j, "Manish"]
print("Original List :",df)
df.remove(89)
df.remove("Ikshit")
df.remove(7-8j)
print("After removal :", df)
```

```python
# pop() function
df = [24, 89, "Ikshit", 7-8j, "Manish"]
print("Original List :",df)
df.pop()
df.pop()
df.pop()
df.pop()
df.pop()
print("After Removal :",df)
print("Current Length of the list: ", len(df))
```

```python
# Slicing in a list:
lst2 = ['P','Y','T','H','O','N']
print(lst2[0:3:1]) #start:stop:step
print(lst2[:3]) # start = 0 (default), step = 1 (default)
print(lst2[1:5:1])
print("Two step position: ",lst2[0:5:2])
print("Entire list: ", lst2[::1])
print("Reversing a list: ", lst2[::-1])
print("Reversing a list with two intervals: ", lst2[::-2])

```
